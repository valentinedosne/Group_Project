```python
class conditions :
    def check_personal_conditon(person):
        total = 0

        if (person["Criminal record"] == True): 
            total += 75

        if (person["Education"] == "Any"): 
            total += 15
        if (person["Education"] == "Undergraduated"):
            total += 5

        if (person["Homeowner"] == False): 
            total += 10

        if (person["Employment"] == "Unemployed"): 
            total += 15
        if (person["Employment"] == "Self-Employed"): 
            total += 5

        if (person["Credit history"] == "Reimbursing"): 
            total += 5
        if (person["Credit history"] == "Less than 2"): 
            total += 15

        return total


    def check_privacy(person): 
        total = 0

        if (person["Marital situation"] == "Single"): 
            total += 5

        if (person["Children"] > 2): 
            total += 10
        if (person["Children"] == 1) or (person["Children"] == 2): 
            total += 5

        if (person["Age"] > 18) and (person["Age"] < 25): 
            total += 2
        if (person["Age"] > 25) and (person["Age"] < 45): 
            total += 5
        if (person["Age"] > 45) and (person["Age"] < 65): 
            total += 10
        if (person["Age"] > 65): 
            total += 25

        return total


    def check_income(person): 

        total = 0    
        if (person["Income"] < 15): 
            total += 12
        if (person["Income"] > 15) and (person["Income"] < 20): 
            total += 9
        if (person["Income"] > 20) and (person["Income"] < 30): 
            total += 6
        if (person["Income"] > 30) and (person["Income"] < 40): 
            total += 3

        if (person["Marital situation"] == "Maried"): 
            if (person["Co-Applicant Income"] < 15): 
                total += 6
            if (person["Co-Applicant Income"] > 15) and (person["Co-Applicant Income"] < 20): 
                total += 4
            if (person["Co-Applicant Income"] > 20) and (person["Co-Applicant Income"] < 30): 
                total += 3
            if (person["Co-Applicant Income"] > 30) and (person["Co-Applicant Income"] < 40): 
                total += 1

        return total
```
